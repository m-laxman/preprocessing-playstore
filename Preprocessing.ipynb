Aim : To perform preprocessing on playstore dataset

```python
import pandas as pd
import numpy as np
```

Read The Data 

```python
df = pd.read_csv('googleplaystore.csv')
```

DATA CLEANING

Count number of Missing values

```python
df.isnull().sum()
```

```python
df
```

Checking for Outliers

```python

df[df.Rating > 5]
```

Eliminating Outliers

```python

df.drop([10472],inplace=True)
```

```python
Data Imputation
fill the null values with appropriate values using aggregate functions such as mean, median or mode.Here We Used Median
```

```python
#Define a function impute_median
def impute_median(series):
    return series.fillna(series.median())
```

```python

df.Rating = df['Rating'].transform(impute_median) 
```

```python
df.isnull().sum()
```

```python
df
```

```python
# modes of categorical values
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

 Fill the missing categorical values with mode

```python

df['Type'].fillna(str(df['Type'].mode().values[0]), inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]), inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]), inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

 Let's convert Price, Reviews and Ratings into Numerical Values

```python

df['Price'] = df['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
df['Price'] = df['Price'].apply(lambda x: float(x))
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: float(x))
```

```python
df
```

Eliminating Duplicates of App Names

```python

df=df.drop_duplicates(subset=['App'])
```

Checking for Any Duplicates

```python

df.App.duplicated().sum()
```
